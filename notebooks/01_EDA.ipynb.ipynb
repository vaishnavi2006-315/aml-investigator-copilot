```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv(
    "../data/raw/LI-Small_Trans.csv",
    nrows=100000
)

df.head()
```

```python
print(df.shape)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Is Laundering"].value_counts()
```

```python
df["Is Laundering"].value_counts().plot(
    kind="pie",
    autopct="%1.2f%%",
    figsize=(6,6)
)
plt.title("Money Laundering Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(df["Amount Paid"], bins=50)

plt.title("Transaction Amount Distribution")

plt.show()
```

```python
df["Payment Format"].value_counts().plot.bar(figsize=(8,4))
plt.show()
```
